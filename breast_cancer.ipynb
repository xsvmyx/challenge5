```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
df = pd.DataFrame(data = data.data, 
                       columns = data.feature_names) 
df
```

```python
df.isna().sum()
```

```python
df.drop_duplicates()
```

**let's try without the PCA**


```python
mat = df.corr().round(2) #2 decimal places
sns.heatmap(data=mat, annot=True)

plt.show()
```

```python
sns.boxplot(data=df)
# Add labels and title
plt.xlabel('Datasets')
plt.xticks(rotation=90)
plt.ylabel('Values')
plt.title('Boxplot')
# Show the plot
plt.show()
```

***Let's find  how many PCA's are the most optimal***


```python
from sklearn.decomposition import PCA

# Appliquer PCA sans réduire le nombre de composants initialement
pca = PCA(n_components=df.shape[1])  
pca.fit(df)

# Calculer la variance expliquée cumulée
explained_variance = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, df.shape[1] + 1), explained_variance, marker='o', linestyle='--')
plt.xlabel('Nombre de composantes principales')
plt.xticks(np.arange(1, df.shape[1] + 1, 1),rotation=90)
plt.ylabel('Variance expliquée cumulée')
plt.title('Méthode du coude pour choisir le nombre de PCA')
plt.grid(True)
plt.show()
```

***2 PCA's are the best choice:***

```python
pca = PCA(n_components=2) 
old_df=df
transformed = pca.fit_transform(df)
df = pd.DataFrame(transformed,columns=["0","1"])
df
```

```python

# Create a boxplot
sns.boxplot(data=df)
# Add labels and title
plt.xlabel('Datasets')
plt.xticks(rotation=90)
plt.ylabel('Values')
plt.title('Boxplot')
# Show the plot
plt.show()
```
